```python
import pandas as pd
import numpy as np
import json
```

```python
df = pd.read_csv('country_vaccinations.csv')
```

```python
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].fillna(0)
```

```python
df.loc[(df["country"]=="Wales")|(df["country"]=="Scotland")]
```

```python
df.loc[(df["country"]=="Wales")|(df["country"]=="Scotland")] = df.loc[(df["country"]=="Wales")|(df["country"]=="Scotland")].fillna('GBR')
```

```python
df.loc[(df["country"]=="Wales")|(df["country"]=="Scotland")]
```

```python
wales = max(df.loc[(df["country"]=="Wales")]['people_fully_vaccinated'].values)
scotland = max(df.loc[(df["country"]=="Scotland")]['people_fully_vaccinated'].values)
```

```python
scotland
```

```python
countries = df['iso_code'].value_counts().keys().values
totVac=[]
```

```python
countries
```

```python
for c in countries:
    if c=='GBR':
        val = max(df.loc[(df['iso_code']==c)]['people_fully_vaccinated'].values) +wales +scotland
        totVac.append({"code3":c, "z":val})
    else:
        val = max(df.loc[(df['iso_code']==c)]['people_fully_vaccinated'].values)
        totVac.append({"code3":c, "z":val})
```

```python
totVac
```

```python
with open('data.json', 'w') as outfile:
    json.dump(totVac, outfile)
```
